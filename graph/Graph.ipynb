```python
import networkx as nx
import pandas as pd
G = nx.Graph()
```

```python
df = pd.read_csv('page.csv', encoding='latin1', header=0)
type(df['origin_link'])
G.add_nodes_from(df['origin_link'])
print(df['origin_link'].nunique())#count unique, matches nodes
print("nodes " + str(G.number_of_nodes()))
G.add_edges_from(tuple(x) for x in df.values)
print("edges " + str(G.number_of_edges()))
df
```

```python
pr = nx.pagerank(G)
```

```python
df_pr = pd.DataFrame(list(pr.items()), columns=['page', 'rank'])
```

```python
df_pr.sort_values('rank', ascending=False)
```
